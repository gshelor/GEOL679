### Q1. Create two `numpy` `arrays`, both of length 8. One array should be all zeros, the second should be all ones.

### Q2. Create a `numpy` `array` with values from 2 through 12 in steps of one, e.g. `[2,3,4,...,11,12]`. Make this `dtype = int`.

```python
array3 = np.arange(2,13, 1, dtype= 'int')
print(array3, array3.dtype)
```

### Q3. Create a `numpy` `array` of 361 values equally spaced between 0 and 2$\pi$. Name this array `x`.

```python
x = np.linspace(0, (2 * np.pi), 361)
print(x)
```

### Q4. Calculate `y1=np.cos(x)` and `y2=np.sin(x)`, where `x` is the array you created above. Plot `x` vs `y`, with `x` on the x-axis and `y1` and `y2` on the y-axis. Add a legend to plot.

```python
y1 = np.cos(x)
y2 = np.sin(x)
plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')
plt.legend(loc= 'best');
```

### Q5. Create an array named `q5arr` with 10 equally spaced integers with the first value 5 and the last value 50. Use slicing to view two new arrays, one with the first 5 values named `q5arr1` and one with the last 5 values named `q5arr2`. Print both of these slices.

```python
q5arr = np.linspace(5,50,10)
q5arr1 = q5arr[0:5]
print(q5arr1)
q5arr2 = q5arr[5:]
print(q5arr2)
```

### Q6. `Reshape` the array `q5arr` to 5x2 and print it, then `reshape` to 2x5 and print again.

```python
q6arr1 = q5arr.reshape(5,2)
print(q6arr1)
q6arr2 = q5arr.reshape(2,5)
print(q6arr2)
```

### Q7. Make arrays `q7arr1` and `q7arr2` as follows and print the resulting arrays:

* `q7arr2` multiplies the values of `q7arr1` and `q5arr2` and adds 3 to the result

### Q8. Make an array that `concatenates` `q7arr1` and `q7arr2`

### Q9. Use logical (boolean) indexing to print the names of the male ducks from Disney movies. Use the data in the cell below, which includes the names, studios (Disney or Warner) and gender identity of 7 classic cartoon characters.

```python
# Use this input data.
names = np.array(["Donald", "Daisy", "Huey", "Dewey", 
                  "Louie", "Daffy", "Daphne"])
studios = np.array(['Disney', 'Disney', 'Disney', 'Disney',
                     'Disney', 'Warner', 'Warner'])
genders = np.array(['male', 'female', 'male', 'male',
                    'male', 'male', 'female'])
```
