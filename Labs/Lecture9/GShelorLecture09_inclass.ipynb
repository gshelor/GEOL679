# In-class Exercises: Matplotlib

```python
import math
import matplotlib.pyplot as plt
```

### We'll use this data for the exercises today:

```python
x = list(range(0,361,4))
y1 = []
y2 = []
for i in range(len(x)):
    y1.append(math.sin(2*x[i]*math.pi/180))
    y2.append(math.cos(3*x[i]*math.pi/180))
```

### Q1. Plot `x` vs `y1` and `x` vs `y2` on the same plot
* `x` should be on the x-axis
* `x` vs `y1` should have a `linewidth` of 2 and a `color` of 'darkgreen'
* `x` vs `y2` should have a `linewidth` of 1 and a `color` of 'hotpink'

```python
# two lines
plt.plot(x, y1, linewidth = 2, color = 'darkgreen')
plt.plot(x, y2, linewidth = 1, color = 'hotpink');
```

### Q2. Plot `x` vs `y1` and `x` vs `y2` on the same plot
* `x` should be on the x-axis
* `x` vs `y1` should have red star markers only (no connecting line)
* `x` vs `y2` should have blue circle markers only (no connecting line)
* Label the x and y axes
* Invert the x axis
* Add a `legend`

```python
# two lines
plt.plot(x, y1, '*', color = 'r')
plt.plot(x, y2, 'o', color = 'b');
```

### Q3. Make 2 subplots that are side-by-side. On the left subplot plot `x` vs `y1` as a green dotted line, and on the right subplot plot as `x` vs `y2` as a blue solid line. Adjust the figure size if needed so that the axis annotations do not overlap with the other subplot.

```python
# figure setup
fig=plt.figure()
ax1=fig.add_subplot(1,2,1)
ax1.plot(x, y1, linestyle = 'dotted', color = 'green')
ax2=fig.add_subplot(1,2,2)
ax2.plot(x, y2, linestyle = 'solid', color = 'blue');
```

### Q4. Make two subplots that are top and bottom. On the top subplot plot `x` vs `y1` with a red dashed line and cyan circle markers, and on the bottom subplot plot as `x` vs `y2` as a blue dotted line and green square markers. Make the figure size (10,10).

```python
# figure setup
fig=plt.figure(figsize=(10, 10))
ax1=fig.add_subplot(2,1,1)
ax1.plot(x, y1, marker = 'o', linestyle = 'dashed', color = 'red', mfc = 'cyan')
ax2=fig.add_subplot(2,1,2)
ax2.plot(x, y2, marker = 's', linestyle = 'dotted', color = 'blue', mfc = 'green');
```

### Q5. Make the same plot as for Q2, but use 2 different matplotlib `styles`.

```python
plt.style.use('ggplot')
plt.plot(x, y1, '*', color = 'r')
plt.plot(x, y2, 'o', color = 'b');
```

```python
plt.style.use('fivethirtyeight')
plt.plot(x, y1, '*', color = 'r')
plt.plot(x, y2, 'o', color = 'b');
```
