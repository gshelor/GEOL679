# In-class Exercises: Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
# Set defaults for plotting (feel free to adjust)
seaborn.set_style("darkgrid")
seaborn.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 2.0})
```

### Q1. Load the dataset 'mpg' from seaborn the using `sns.load_dataset()`. Display the `DataFrame`.

```python
mpg_data = seaborn.load_dataset('mpg')
mpg_data
```

### Q2. Make a `seaborn` scatter plot with `relplot()`, plotting 'mpg' on the y-axis versus 'weight' on the x-axis. 

```python
## scatter plot
seaborn.relplot(x = 'weight', y = 'mpg', data = mpg_data);
```

### Q3. Fit a standard, least-squares regression model to the data. Change the default line color with the argument `line_kws={"color": "XXX"}`.

```python
# Linear regression
seaborn.regplot(x="weight", y="mpg", data= mpg_data, line_kws = {"color": "gold"});
```

### Q4. Using another line color, fit a robust linear regression model to the data.

```python
seaborn.lmplot(x="weight", y = "mpg", data = mpg_data, robust=True, ci=None,
    line_kws={"color": "orange"});
```

### Q5. Using another line color, fit a quadratic regression model to the data.

```python
# Linear model, quadratic polynomial
seaborn.regplot(x="weight", y="mpg", order=2, data=mpg_data, line_kws={"color": "red"});
```

### Q6. Make a `PairGrid` of 'mpg', 'weight', 'displacement', and 'horsepower'.

Note: you can select the columns of interest with the argument `vars=[...]`).

```python
# Basic PairGrid
g = seaborn.PairGrid(mpg_data, vars=['mpg', 'weight', 'displacement', 'horsepower']) # initialize a blank grid
g.map_diag(seaborn.histplot) # put a histogram on the diagonal
g.map_offdiag(seaborn.scatterplot);
```

### Q7. Make a violin plot of 'mpg' by country of 'origin'. Manually set the x and y labels.

```python
# using seaborn
ax = seaborn.catplot(x="origin", y="mpg", kind="violin", data=mpg_data)
ax.set(xlabel="Country of Origin", ylabel="Miles per Gallon");
```
