```python
import numpy as np
import matplotlib.pyplot as plt
```

### Q1. Load data from the file `lecture11_vel_sedthick.csv` into two `numpy` `arrays`
* The file has one header row that you should skip
* The file uses ',' as a `delimiter`
* First column is seismic velocity in a geologic layer right below the sediment-basement interface in oceanic crust
* Second column is the corresponding sediment thickness above this layer

```python
velocity, sed_thick = np.loadtxt('lecture11_vel_sedthick.csv', delimiter=',', skiprows=1, unpack=True)
```

### Q2. Make a plot with velocity on the x-axis and sediment thickness on the y-axis. Use markers (not a standard line plot) and label the axes.

```python
plt.plot(velocity, sed_thick, 'o')
plt.xlabel('Seismic Velocity')
plt.ylabel('Sediment Thickness');
```

### Q3. Fit the data with a polynomial of an appropriate degree (you decide), and make the same plot as Q2, but add the polynomial fit as a line. Include a legend on the plot.

```python
## fitting polynomial
## sorting data
velocity_sort = np.sort(velocity)
sed_thick_sort = np.sort(sed_thick)
data_fit = np.polyfit(velocity_sort, sed_thick_sort, 3)
y_calc = np.polyval(data_fit, velocity_sort)

plt.plot(velocity, sed_thick, 'o', label = 'data')
plt.plot(velocity_sort, y_calc, "-", label='polynomial fit degree 3')
plt.xlabel('Seismic Velocity')
plt.ylabel('Sediment Thickness')
plt.legend(loc = 'best');
```

### Q4. Use `np.select` to create an `array` that has the value 'thin' if sediment thickness is less than 1 km, 'normal' if sediment thickness is between 1-3.5 km, and 'thick' if sediment thickness is greater than 3.5 km.

```python
## setting conditions for selection
conditions = [(sed_thick < 1), 
              ((sed_thick >= 1) & (sed_thick < 3.5)), 
              (sed_thick >= 3.5)]
## setting categories for each condition
categories = ['thin', 'normal', 'thick']
sed_thick_cat = np.select(conditions, categories)
print(sed_thick_cat)
```

### Q5. What are the mean and standard deviation of sediment thickness measurements? What are the mean and standard deviation of the velocity measurements?

```python
sed_thick_mean = np.mean(sed_thick)
sed_thick_sd = np.std(sed_thick)
print(sed_thick_mean, sed_thick_sd)

velocity_mean = np.mean(velocity)
velocity_sd = np.std(velocity)
print(velocity_mean, velocity_sd)
```

### Q6. Make an `array` that has the velocity sorted by increasing velocity, and a second `array` that has the sediment thicknesses corresponding to these sorted velocity values. 

```python
## sorting data
velocity_sort_ind = np.argsort(velocity)
# print(velocity_sort_ind)
velocity_sort = velocity[velocity_sort_ind]
print(velocity_sort)
sed_thick_sort = sed_thick[velocity_sort_ind]
print(sed_thick_sort)
```
