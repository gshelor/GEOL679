### `np.random()`
* `np.random.rand(n)` - returns an array of n random numbers between 0 and 1 (i.e. a random sample from the Uniform distribution)
* `np.random.randn(n)`  - returns an array of n random numbers centered at 0 with standard deviation 1 (i.e a random sample from the Normal distribution)

```python
# Uniform random numbers in the interval [0,1]
arr_r1 = np.random.rand(5)
arr_r2 = np.random.rand(5)
arr_r3 = np.random.rand(5)
print(arr_r1)
print(arr_r2)
print(arr_r3)
```

```python
# Normally distributed, most values between (-2,2)
arr_r4 = np.random.randn(5)
arr_r5 = np.random.randn(5)
arr_r6 = np.random.randn(5)
print(arr_r4)
print(arr_r5)
print(arr_r6)
```

```python
# It is easy to shift or rescale these distributions
start = 5.0
width = 2.0
arr_r7 = start + width * np.random.rand(5)
print(arr_r7)
```

```python
# It is easy to shift or rescale these distributions
mean = 10.0
sdev = 2.0
arr_r8 = mean + sdev * np.random.randn(5)
print(arr_r8)
```

```python
# Histograms: show the distribution of numbers - Uniform

# one array of random numbers
start1 = 1.0
stop1 = 3.0
arr1 = start1 + (stop1 - start1) * np.random.rand(10000) 

# one array of random numbers
start2 = 5.0
stop2 = 9.0
arr2 = start2 + (stop2 - start2) * np.random.rand(10000) 

# plot their distribution
hbinE = np.arange(0.0, 10.1, 0.2)
plt.hist(arr1, hbinE)
plt.hist(arr2, hbinE)
plt.xlabel("Value")
plt.ylabel("Count")
plt.xticks(range(0, 11));
```

```python
# Histograms: show the distribution of numbers - Normal

# one array of random numbers
mean1 = 15.0
sdev1 = 2.0
arr1 = mean1 + sdev1 * np.random.randn(10000) 

# a second array
mean2 = -15.0
sdev2 = 4.0
arr2 = mean2 + sdev2 * np.random.randn(10000)

# plot their distribution
hbinE = np.arange(-30.0, 25.1, 0.5)
plt.hist(arr1, hbinE)
plt.hist(arr2, hbinE)
plt.xlabel("Value")
plt.ylabel("Count");
```

### `np.where()` - use instead of `for` loops with `if/else` statements
Creates a new array based with two different values depending on whether the condition is True or False. The values can be constants or arrays.
* `arr = np.where(condition, val1, val2)`

```python
# np.where example
arr1 = np.random.randn(5) # initial array
print(arr1)
arr2 = np.where(arr1 < 0, 1, 2) # a new array with values of 1 and 2
print(arr2)
```

```python
# np.where example
arr3 = np.random.randn(5) # initial array
print(arr3)
arr4 = np.where(arr3 < 0, 'Low', 'High') # a new array with values of 'Low' and 'High'
print(arr4)
```

```python
# np.where with arrays as the values
arr6 = np.where(arr2 > 0, arr3, -arr3)
print(arr6)
```

### `np.select()` - like `np.where()` but for more than 2 conditions

```python
arr1 = 10 * np.random.randn(5) # initial array
print(arr1)

# conditions to check
conditions = [(arr1 < -20), 
              ((arr1 >= -20) & (arr1 < 0)), 
              ((arr1 >= 0) & (arr1 < 20)),
              arr1 >= 20] 

choices = [1, 2, 3, 4] # choices for each condition

arr2 = np.select(conditions, choices) # new array
print(arr2)
```

```python
arr3 = 80.0 + 10 * np.random.randn(10) # initial array
print(arr3)

# conditions to check
conditions = [(arr3 > 90), 
              ((arr3 >= 80) & (arr3 < 90)), 
              ((arr3 >= 70) & (arr3 < 80)), 
              ((arr3 >= 60) & (arr3 < 70)), 
              (arr3 < 60)]

choices2 = ['A', 'B', 'C', 'D', 'F'] # choices for each condition

arr4 = np.select(conditions, choices2) # new array       
print(arr4)
```

### Sorting
* `np.sort()` - returns a sorted copy of array

```python
# Plot a random time series
x1 = np.arange(1990, 2000, 1)
y1 = np.array([5, -2, 6, 8, 1, -12, -20, -3, 9, 0])
plt.plot(x1, y1, '*')
plt.xlabel('Year')
plt.ylabel('Some Parameter');
```

```python
# Sorts in ascending order
y1_sort = np.sort(y1)
print(y1_sort)
```

```python
# Sort in descending order
y1_rsort = np.sort(y1)[::-1]
print(y1_rsort)
```

`np.argsort()` - returns indices that would sort an array

```python
# argsort gives indices needed to sort an array
ind2 = np.argsort(y1)
print(y1)
print(ind2)
```

These indices may see counterintuitive at first. The interpretation is as follows. The first number `argsort` returns is 6. That means that to sort `y1`, we need to move the number sitting in `y1` at position 6 to the front. This checks out: `y1[6]` is -20, the smallest number in the array. Similarly, the second number `argsort` returns is 5, meaning that we need to place `y1[5] = -12` in second position to make the sorted array. And so on. 

```python
# See?
x1_sort = x1[ind2]
y1_sort = y1[ind2]
print(x1_sort)
print(y1_sort)
```

### Polynomial fit to data
* First, calculate polynomial coefficients with `c = np.polyfit(x, y, degree)`
* With the coefficients in hand, then evaluate with `np.polyval(c, xval)`.
* Note that `xval` does not have to be the same array as `x`.

#### Generate some interesting arrays

```python
# Don't worry, just generating data
x = 2*(np.random.rand(51)) - 1
y1 = 3*x + 0.2 + 0.5*np.random.randn(len(x)) 
y2 = 4*x**2 + 3*x + 0.2 + 0.5*np.random.randn(len(x)) 
y3 = x**3 - x + 0.1*np.random.randn(len(x)) 
```

```python
# Fitting a line using polyfit and polyval

# step 1: calculate the polynomial fit
c1 = np.polyfit(x, y1, 1) 
print("\npolyfit:", c1)

# step 2: using the polyfit, calculate values for an input array x
xval = x # values to calculate fit
ycalc1 = np.polyval(c1, xval) 
print("\npolyval:", ycalc1)
```

```python
#  Plot fit to data
plt.plot(x, y1,'x', label='values') # plot data
plt.plot(xval, ycalc1, '.', label='polynomial fit degree 1') # plot polynomial
plt.legend(loc='best'); # add a legend
```

```python
# Fit a degree two polynomial

# step 1: calculate the polynomial fit
c2 = np.polyfit(x, y2, 2) 
print("\npolyfit:", c2)

# step 2: using the polyfit, calculate values for an input array x
xval = x # values to calculate fit
ycalc2 = np.polyval(c2, xval) 
print("\npolyval:", ycalc2)
```

```python
# Plot fit to data
plt.plot(x, y2, 'x', label='values') # plot data
plt.plot(xval, ycalc2, '.', label='polynomial fit degree 2') # plot polynomial
plt.legend(loc='best'); # add a legend
```

### Assigning variables to columns in a file with `np.loadtxt()`
* `x, y = np.loadtxt(filename,unpack=True)` for file with 2 columns separated by white space
* `add delimiter=','` if columns are separated by commas
* add `skiprows=n` if you need to skip n rows at the beginning
* add `usecols=(0,2)` if you want to specify columns to use (e.g., 0 and 2)

```python
# load data into two 1D arrays
v1,z1 = np.loadtxt('lecture11_input1.txt', unpack=True)
print(v1)
print(z1)
```

```python
# add options
v2,z2 = np.loadtxt('lecture11_input2.txt', delimiter=',',
                   skiprows=1, usecols=(0,1), unpack=True)
print(v2)
print(z2)
```

```python
# plot results
plt.plot(v1, z1, label='File1')
plt.plot(v2, z2, label='File2')
plt.xlabel('Velocity (km/s)', fontsize=14)
plt.ylabel('Depth (km)', fontsize=14)
plt.legend(loc='lower left', fontsize=12)
plt.tick_params(labelsize=12)
plt.gca().invert_yaxis();
```
