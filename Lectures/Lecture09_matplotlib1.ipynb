# Matplotlib: 

### Matplotlib is a Python's standard plotting library which produces publication quality figures in a variety of output formats and interactive environments across platforms. https://matplotlib.org/index.html

### It is recommended to `import matplotlib.pyplot as plt` 

(this standard nickname helps others use your code).

```python
import math
import matplotlib.pyplot as plt
```

### Let's make some interesting data to plot

```python
# example: a sine curve
x = list(range(0, 361, 12))
y1 = []
for i in range(len(x)):
    y1.append(math.sin(x[i]*math.pi/180))
y2 = []
for i in range(len(y1)):
    y2.append(y1[i] + x[i]*0.005)
```

### A simple example

```python
# simplest plot
plt.plot(x, y1)
```

### Add a `;` after the last command if you don't want the matplotlib text information displayed, or use `plt.show()`

```python
# identical except the ;
plt.plot(x, y1);
```

```python
# plt.show() does same
plt.plot(x, y1)
plt.show()
```

### Multiple lines on the same plot

```python
# two lines
plt.plot(x, y1)
plt.plot(x, y2);
```

### Change the font size

Defaults for plotting are controlled by `plt.rcParams` (technically, a Python dictionary). There are many options: https://matplotlib.org/stable/api/matplotlib_configuration_api.html#matplotlib.RcParams.

Note that if you make a change to `plt.rcParams`, this change will remain in place until you restart your notebook (or overwrite it).

```python
# Example: change default fontsize
plt.rcParams['font.size'] = 14  
plt.plot(x, y1)
plt.plot(x, y2);
```

```python
# There are a load of options!!!
plt.rcParams
```

### Often, we want to control the details of individual plots.
#### For example, if we want to change the line `color`:

```python
# color change
plt.plot(x, y1, color='orange')
plt.plot(x, y2, color='purple');
```

#### Some hotkeys for colors: 'b'=blue, 'r'=red, 'y'=yellow, 'g'=green, 'k'=black, 'm'=magenta, 'c'=cyan

```python
# color hotkeys
plt.plot(x, y1, 'b')
plt.plot(x, y2, 'r');
```

#### Can get any color on the spectrum by specifying its hex code

```python
# hex colors: UNR color scheme
plt.plot(x, y1, color="#003366"); # "UNR primary blue"
plt.plot(x, y2, color="#373E45") # "UNR dark gray"
```

#### Or using the many named colors (https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
# named colors
plt.plot(x, y1, color='dodgerblue')
plt.plot(x, y2, color='darkgreen');
```

#### Changing `linewidth`

```python
# linewidths
plt.plot(x, y1, linewidth=4.0)
plt.plot(x, y2, linewidth=0.25)
plt.show()
```

#### Changing `linestyle`

```python
# compare styles
plt.plot(x, y1, linestyle='dotted')
plt.plot(x, y2, linestyle='dashed')
plt.show()
```

```python
# shorthand
plt.plot(x, y1, ':')
plt.plot(x, y2, '--');
```

```python
# customized (dash and space widths)
plt.plot(x, y1, dashes=[6, 4])
plt.plot(x, y2, dashes=[10, 4, 2, 4]);
```

#### Plots with markers instead of lines or in addition to lines

```python
# markers linestyle
plt.plot(x, y1, 's') # s=square
plt.plot(x, y2, 'o'); # o=circle
```

```python
# use "marker" command to have original line included
plt.plot(x, y1, marker='x')
plt.plot(x, y2, marker='*');
```

### more customization

```python
# we can tweak many things
plt.plot(x, y1, 's', markerfacecolor='green',
         markeredgecolor='blue', markersize=8.0, markeredgewidth=1.5)
plt.plot(x, y2, 'o', markerfacecolor='purple',
         markeredgecolor='orange', markersize=8.0, markeredgewidth=1.5);
```

```python
# we can use shorthand
plt.plot(x, y1, 's', mfc='green', mec='blue', ms=8.0, mew=1.5)
plt.plot(x, y2, 'o', mfc='purple', mec='orange', ms=8.0, mew=1.5);
```

### `legend`

```python
# legend example
plt.plot(x, y1, label='line1')
plt.plot(x, y2, label='line2')
plt.legend(loc = 'best'); # automatically locates where there is the most space
```

```python
# adding labels
plt.plot(x, y1, 's', mfc='green', mec='blue', 
         ms=8.0, mew=1.5, label="green squares")
plt.plot(x, y2,'o',mfc='purple',mec='orange', 
         ms=8.0, mew=1.5, label="purple circles");
plt.legend(loc='best');
```

#### Axis `limits` and `labels`

```python
# setting limits and labels
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim([0, 300])
plt.xlabel('Degrees')
plt.ylim([-1, 2])
plt.ylabel('Value');
```

#### Axis `ticks`

```python
# setting ticks
plt.plot(x,y1)
plt.plot(x,y2)
plt.xticks([0, 90, 180, 270, 360])
plt.yticks([-1, 0, 1, 2])
plt.xlim([0, 360])
plt.ylim([-1, 2])
plt.xlabel('Degrees')
plt.ylabel('Value');
```

```python
# setting ticks
plt.plot(x, y1)
plt.plot(x, y2)
plt.xticks(range(0, 361, 90)) # can use range
plt.yticks(range(-1, 3)) # can use range
plt.xlim([0, 360])
plt.ylim([-1, 2])
plt.grid(lw=0.25) # add a grid
plt.xlabel('Degrees')
plt.ylabel('Value');
```

#### Axis `tick_params`

```python
# tick_params
plt.plot(x, y1)
plt.plot(x, y2)
plt.xticks(range(0, 361, 90)) # can use range
plt.yticks(range(-1, 3)) # can use range
plt.xlim([0, 360])
plt.ylim([-1, 2])
plt.grid(lw=0.5) # add a grid
plt.xlabel('Degrees')
plt.ylabel('Value')
plt.tick_params(axis='both', length=10.0, width=1.5, rotation=30); # lots of customization!
```

### Inverting an axis

```python
# x-axis reads backwards
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim([0, 360])
plt.xlabel('Degrees')
plt.ylabel('Value')
plt.gca().invert_xaxis(); # Invert x-axis
```

```python
# y-axis reads backwards
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlim([0, 360])
plt.xlabel('Degrees')
plt.ylabel('Value')
plt.gca().invert_yaxis(); # Invert y-axis
```

### Working with log plots

```python
# logscale for x and y
xlog = [1, 10, 100, 1000, 10000]
ylog = [2, 25, 400, 3000, 15000]
plt.loglog(xlog, ylog, 's')
plt.loglog(xlog, ylog, '-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('loglog plot');
```

```python
# logscale for x only
plt.semilogx(xlog, ylog, 's')
plt.semilogx(xlog, ylog, '-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('semilogx plot');
```

```python
# logscale for y only
plt.semilogy(xlog, ylog, 's')
plt.semilogy(xlog, ylog, '-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('semilogy plot');
```

### Subplots - multiple plots
* basic syntax is `fig.add_subplot(nrow, ncol, number)`

```python
# figure setup
fig=plt.figure(figsize=(15,7))

# create subplots on a 2x3 grid
ax1=fig.add_subplot(2,3,1)
ax1.text(0.5, 0.5, '(2,3,1)', fontsize=24, horizontalalignment='center')
ax2=fig.add_subplot(2,3,2)
ax2.text(0.5, 0.5, '(2,3,2)', fontsize=24, horizontalalignment='center')
ax3=fig.add_subplot(2,3,3)
ax3.text(0.5, 0.5,'(2,3,3)', fontsize=24, horizontalalignment='center')
ax4=fig.add_subplot(2,3,4)
ax4.text(0.5, 0.5,'(2,3,4)', fontsize=24, horizontalalignment='center')
ax5=fig.add_subplot(2,3,5)
ax5.text(0.5, 0.5,'(2,3,5)', fontsize=24, horizontalalignment='center')
ax6=fig.add_subplot(2,3,6);
ax6.text(0.5, 0.5,'(2,3,6)', fontsize=24, horizontalalignment='center');
```

```python
# figure setup
fig=plt.figure()

# 2x2 grid of subplots
ax1=fig.add_subplot(2,2,1)
ax1.plot(x, y1, 'r')
ax2=fig.add_subplot(2,2,2)
ax2.plot(x, y1, 'b')
ax3=fig.add_subplot(2,2,3)
ax3.plot(x, y1, 'g')
ax4=fig.add_subplot(2,2,4)
ax4.plot(x, y1, 'c');
```

```python
# we don't need to add all panels
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.plot(x, y1, 'r')
ax4=fig.add_subplot(2,2,4)
ax4.plot(x, y1, 'c');
```

### I think this syntax for subplots is simpler/cleaner:

```python
# I like this syntax better
fig, axi = plt.subplots(2, 2) # two rows, two columns
axi[0,0].plot(x, y1, 'r') # row 0, column 0
axi[1,0].plot(x, y1, 'b') # row 1, column 0
axi[0,1].plot(x, y1, 'g') # row 0, column 1
axi[1,1].plot(x, y1, 'c'); # row 1, column 1
```

```python
# 1 x 2 plot (indexing a 1d array): notice difference in indexing
fig, axi = plt.subplots(1, 2, figsize=(5, 2.5)) # one row, two columns
axi[0].plot(x, y1, 'r') #column 0
axi[1].plot(x, y1, 'g'); # column 1
```

For subplots, we set `labels`, `ticks`, etc. at the *axis* level instead of *figure* level:
* `axi[i,j].set_xlim(...)`
* `axi[i,j].set_xticks(...)`
* `axi[i,j].set_xlabel(...)`

rather than:

* `plt.xlim(...)`
* `plt.xticks(...)`
* `plt.xlabel(...)`

```python
# example: setting labels

# figure setup
fig, axi = plt.subplots(2,2, sharex=True, sharey=True) # two rows, two columns
fig.subplots_adjust(hspace=0.1, wspace=0.1) # adjusts spacing

# plot data
axi[0,0].plot(x, y1, 'r') # row 0, column 0
axi[1,0].plot(x, y1, 'b') # row 1, column 0
axi[0,1].plot(x, y1, 'g') # row 0, column 1
axi[1,1].plot(x, y1, 'c') # row 1, column 1

# add labels and ticks
axi[0,0].set_ylabel("data", fontsize=14)
axi[1,0].set_ylabel("data", fontsize=14)
axi[1,0].set_xlabel("time", fontsize=14)
axi[1,1].set_xlabel("time", fontsize=14)
axi[0,0].set_yticks([-1, -0.5, 0, 0.5, 1])
axi[1,0].set_yticks([-1, -0.5, 0, 0.5, 1])
axi[1,0].set_xticks(range(0, 400, 50))
axi[1,1].set_xticks(range(0, 400, 50));
```
