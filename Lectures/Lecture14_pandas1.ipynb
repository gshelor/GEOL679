## Pandas - Python Data Analysis Library 

Pandas is designed to make working with structured or tabular data (e.g., spreadsheets) fast and easy. I find it to be one of the most useful Python packages, and import it in nearly every notebook I write. There are many features that are described in the documentation (https://pandas.pydata.org/), and we'll only get to cover a sampling over the next three lectures.

```python
import numpy as np
import pandas as pd # this is the standard import
import matplotlib.pyplot as plt
```

### Pandas `DataFrame`: these store data tables that contain a set of columns which can be of different value types (numeric, string, etc), but are all the same length. 

* All `DataFrames` have both row and column indices.
* The rows are usually different data points.
* The columns are usually different data attributes for each data point.

### We can access a column by its column label, a bit like using a key in a Python dictionary. 

This produces a pandas `Series`, which is similar to a one-dimensional NumPy `array` except it includes an index to match all values.

```python
# Series
age_values = df["Age"] # extract Age as a series
print(type(age_values))
age_values # indices on the left, values on the right
```

```python
# Or with "dot notation"
age_values = df.Age
age_values # indices on the left, values on the right
```

### What if we want just the `values` as an `numpy` array? Simple!

```python
# numpy arrays
age_values2 = df["Age"].values
print(len(age_values2), type(age_values2))
print(age_values2)
```

### We can access multiple columns by putting column names in a list. This will produce a `DataFrame` (this is a view, not a copy).

### We can access rows or columns by position (numerical index) using `.iloc`. This is like slicing a 2D NumPy `array`.

```python
# iloc
df.iloc[:5, 1:3] # first five rows, columns 1-3
```

```python
df.iloc[:5] # all columns, first 5 rows
```

```python
df.iloc[:, :3].head() # header, three columns
```

### Using `.loc` is similar to `.iloc`, but using the column names directly.

**A key difference**: the row and column indexing is inclusive!

```python
df.loc[:5,["Name", "Sex", "Age"]] # three columns
```

```python
df.loc[:6, "Height":] # all columns after height
```

### We can get subsets of data using logical / conditional slices

```python
# Only gold medal data
df_gold = df[df["Medal"] == 'Gold'] # df["Medal"] == "Gold" is a logical slice
df_gold.head()
```

```python
# The preferred syntax, using loc
df_gold = df.loc[df["Medal"] == 'Gold', :] # select rows that meet a condition
df_gold.head()
```

```python
# Select columns, only gold medal winners
df_gold_height = df.loc[df["Medal"] == 'Gold', ["Name", "Team", "Year"]]
df_gold_height.head()
```

```python
# Select gold or silver winners
df_goldsilver = df.loc[(df["Medal"] == 'Gold') | (df["Medal"] == 'Silver'), :]
df_goldsilver.head()
```

```python
# Select columns in addition to this
df_goldsilver_weight = df.loc[(df["Medal"] == 'Gold') | (df["Medal"] == 'Silver'), ["Height", "Weight"]]
df_goldsilver_weight.head()
```

```python
# Plot and compare to full dataframe
plt.plot(df["Height"], df["Weight"], 'kx', label='all')
plt.plot(df_goldsilver["Height"], df_goldsilver["Weight"], "s",
         color="red", markersize=5.0, label='Gold/Silver')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend(loc='best');
```

* `.sum()`

### We can delete a column using `del`

```python
del df["Event"] # removes the column
df.head()
```

### It is easy to add a new column

```python
df["BMI"] = (10000.0 * df["Weight"]) / (df["Height"] * df["Height"]) # body mass index
df.head()
```

### Often we want to sort a file by column values. Here we use `sort_values`, which by default does _not_ operate in place and instead returns a copy.

```python
# Sort by year
df_gold_sort = df_gold.sort_values(by=["Year"]) # returns a copy
df_gold_sort.head(10)
```

```python
# Reverse sort by year
df_gold_sort2 = df_gold.sort_values(by=["Year"], ascending=False)
df_gold_sort2.head(10)
```

```python
# Sort full database by year then team
df_sort = df.sort_values(by=["Year", "Team"], ascending=[False, True])
df_sort.head(10)
```

### We can write out a csv file using `.to_csv(filename)`

```python
df.to_csv('lecture14_athletes_men400m_bmi.csv', index=False) # no index column on output
```

### We can make DataFrame from other objects, like `lists`, numpy `arrays`, and `dictionaries`.

```python
# list of name, major, score 
name = ["Aparna", "Junlin", "Sally", "Jeff"] 
degree = ["geology", "biology", "accounting", "chemistry"] 
score = [90, 85, 80, 98] 
```

```python
# Dataframe from lists or numpy arrays
df = pd.DataFrame()
df["Name"] = name
df["Degree"] = degree
df["Score"] = score
df
```

```python
# dictionary of lists (column names are keys, column data are values)
dict = {'Name': name, 'Degree': degree, 'Score': score}  
df = pd.DataFrame(dict) 
df
```

# Summary
* `pandas` is a powerful module for working with tabular data.
* The core data type is a `DataFrame` with columns that organize arrays of data (like a spreadsheet).
* `DataFrames` are a way of organizing `Series` of data, which are like `numpy` arrays with indices corresponding to each row.
* We can slice `DataFrames` with `.loc` or `.iloc`. This is useful to get subsets of data.
* `DataFrames` have built-in statistical and plotting functions.
* We can also sort `DataFrames` by column values.
* Reading and writing csv files is straightforwad: `read_csv()` and `to_csv()`.
* We can also create a `DataFrame` from scratch, given lists, arrays, or dictionaries of data.
